```python
import os, json, subprocess, csv, math, networkx as nx
from auguria import grid, gridlab
from matplotlib import pyplot as plt
```

```python
xFeed=grid.parse('testVolt.glm')
```

```python
grid.latLonNxGraph(grid.treeToNxGraph(xFeed))
```

```python
# Run Gridlab. First write GLM:
with open('testVolt.glm','w') as glmFile:
	glmFile.write(grid.sortedWrite(xFeed))
# Then make the magic happen:
with open('stdout.txt','w') as stdout, open('stderr.txt','w') as stderr, open('PID.txt','w') as pidFile:
    proc = subprocess.Popen(['gridlabd','-w','testVolt.glm'], cwd='.', stdout=stdout, stderr=stderr)
    pidFile.write(str(proc.pid))
returnCode = proc.wait()
os.remove('PID.txt')
```

```python
# Gridlab profile results to make sure we ran well.
with open('stdout.txt','r') as inFile:
    print inFile.read()
```

```python
# Pull in the results:
with open('voltDump.csv','r') as dumpFile:
    reader = csv.reader(dumpFile)
    print 'Headers:', reader.next()
    keys = reader.next()
    print 'Keys:', keys
    voltTable = []
    for row in reader:
        rowDict = {}
        for pos,key in enumerate(keys):
            rowDict[key] = row[pos]
        voltTable.append(rowDict)
# View results:
print 'Example row:', voltTable[0]
```

```python
# Calculate average node voltage deviation. First, helper functions.
def pythag(x,y):
    ''' For right triangle with sides a and b, return the hypotenuse. '''
    return math.sqrt(x**2+y**2)
def digits(x):
    ''' Returns number of digits before the decimal in the float x. '''
    return ceil(log10(x+1))
def avg(l):
    ''' Average of a list of ints or floats. '''
    return sum(l)/len(l)
# Detect the feeder nominal voltage:
for key in xFeed:
    ob = xFeed[key]
    if ob.get('bustype','')=='SWING':
        feedVoltage = float(ob.get('nominal_voltage',1))
# Tot it all up.
nodeVolts = {}
for row in voltTable:
    allVolts = []
    for phase in ['A','B','C']:
        phaseVolt = pythag(float(row['volt'+phase+'_real']),
                           float(row['volt'+phase+'_imag']))
        if phaseVolt != 0.0:
            if digits(phaseVolt)>3:
                # Normalize to 120 V standard
                phaseVolt = phaseVolt*(120/feedVoltage)
            allVolts.append(phaseVolt)
    nodeVolts[row.get('node_name','')] = avg(allVolts)
# Examples.
print 'Example Deviations:'
for key in nodeVolts.keys()[0:5]:
    print key, nodeVolts[key]
```

```python
# Color nodes by VOLTAGE.
fGraph = grid.treeToNxGraph(xFeed)
plt.figure(figsize=(12,10))
plt.axes(frameon = 0)
plt.axis('off')
positions = {n:fGraph.node[n].get('pos',(0,0)) for n in fGraph}
edgeIm = nx.draw_networkx_edges(fGraph, positions)
nodeIm = nx.draw_networkx_nodes(fGraph,
    pos = positions,
    node_color = [nodeVolts.get(n,0) for n in fGraph.nodes()],
    linewidths = 0,
    node_size = 30,
    cmap = plt.cm.jet)
plt.sci(nodeIm)
plt.clim(110,130)
plt.colorbar()
plt.show()
```
