```python
# Import CSV file.
with open('demandResponseData.csv','r') as inFile:
    fullData = inFile.read()
# Clean and split string.
fullDataSplit = [x.split(',') for x in fullData.split('\n')]
headers = fullDataSplit[0]
body = fullDataSplit[1:]
print 'header:', headers
print 'bodyExample:', body[1:3]
# Turn arr-arr into arr-dict.
niceData = [{'dateTime':line[0],'demand':float(line[1]),'step':int(line[2])} for line in body]
print 'arr-dictExample:', niceData[1:3]
```

```python
# Define best fit functions
def r2(x,y):
    ''' Return the r^2 statistic for the best fit to a given set of points. '''
    model, resid = np.polyfit(array(x),array(y),1,full=True)[:2]
    r2Value = 1 - resid / (array(y).size * array(y).var())
    return r2Value[0]
def mxpb(x,y):
    ''' Return m and b for best fit line. '''
    model, resid = np.polyfit(array(x),array(y),1,full=True)[:2]
    return model[0], model[1]
```

```python
def r2ArrayGen(inArray):
    ''' Array showing probability of data inflexion point. '''
    r2Array = []
    for breakPoint in xrange(len(inArray)):
        # Ignore 5 points on both sides of time series.
        if breakPoint<5 or breakPoint>len(inArray)-6:
            r2Array.append(0)
        else:
            firstPart = inArray[0:breakPoint]
            secondPart = inArray[breakPoint:]
            rScore = r2(range(len(firstPart)), firstPart) * \
                     r2(range(len(secondPart)),secondPart)
            r2Array.append(rScore)
    return r2Array
def maxIndex(inR2Array):
    ''' Index of first max in array data. '''
    return inR2Array.index(max(inR2Array))
# Test on our DR data, starting with just plotting it.
allDemands = [x['demand'] for x in niceData]
drR2Scores = r2ArrayGen(allDemands)
drSplit = maxIndex(drR2Scores)
fig, ax1 = plt.subplots(figsize=[12,5])
ax1.plot(allDemands)
# Best fit lines.
prePeak=allDemands[0:drSplit]
preModel = mxpb(range(len(prePeak)),prePeak)
ax1.plot(range(drSplit), preModel[0]*array(range(drSplit))+preModel[1], 'b--')
postPeak = allDemands[drSplit:]
postRange = drSplit+np.arange(len(postPeak))
postModel = mxpb(postRange,postPeak)
ax1.plot(postRange, postModel[0]*postRange+postModel[1], 'b--')
# R2 split scores.
ax2 = ax1.twinx()
ax2.plot(drR2Scores, 'g.')
ax2.axvline(drSplit, color='red')
plt.show()
```
